```python
import requests
#url = "https://www.ens-lyon.fr/LBMC"
url = "http://www.ebi.ac.uk/Tools/webservices/psicquic/intact/webservices/current/search/query/pubid:17446270"

try:
    httpReq = requests.get(url, proxies=None)
except NameError:
    httpReq = requests.get(url)
ans = httpReq.text
```

```python
import re

def mitabReader(httpText):
    for line in ans.split("\n"):
        _ = line.split("\t")
        if len(_) > 1 and _[0].startswith("uniprotkb:")\
                      and _[1].startswith("uniprotkb:"):
            yield [ _[0].replace("uniprotkb:", ""),\
                    _[1].replace("uniprotkb:", "") ]\
                  + _[2:]
```

```python
mitabReader(ans) #return un objet de type generateur
#returner une valeur pour chaque valeur du boucle (yield)
#pas une liste car on connais pas la taille du liste
t = mitabReader(ans)
#for l in t:
    #print(l)

#for l in t:
    #print(l)
    #it is exhausted (tourne pas une 2e fois)
```

```python
mitab_list = [ d for d in mitabReader(ans) ]

#we need to split the mitab_list into 2-specie-specific lists
#2 lists --> ebv_ebv_mitab_list, humaine_ebv_mitab_list
```

```python
import re

def isMitab_EBV_EBV(mitabLine):
    reEBV   = "taxid:(1037[6-7]|82830)"
    if re.search(reEBV, mitabLine[9]) and re.search(reEBV, mitabLine[10]):
        return True
    return False

def isMitab_Human_EBV(mitabLine):
    #human match only
    #reEBV   = "taxid:(1037[6-7]|82830)"
    reHUM   = "taxid:9606"
    if re.search(reHUM, mitabLine[9]) or re.search(reHUM, mitabLine[10]):
        return True
    return False


EBV_EBV_mitab   = []
EBV_Human_mitab = []
total = 0
for mitabArray in mitabReader(ans):
    total += 1
    if isMitab_EBV_EBV(mitabArray):
        EBV_EBV_mitab.append(mitabArray)
    elif isMitab_Human_EBV(mitabArray):
        EBV_Human_mitab.append(mitabArray)
    else : 
        raise ValueError("Je ne connais pas cette espece ==> ", mitabArray[9:11])

print(f"Nombre total d'interactions {total}, EBV-EBV {len(EBV_EBV_mitab)} , EBV - HUM {len(EBV_Human_mitab)}")
```

```python
EBV_prot = set()
for mitab in EBV_EBV_mitab:
    EBV_prot.add(mitab[0])
    EBV_prot.add(mitab[1])

EBV_HUM_prot = set()
for mitab in EBV_Human_mitab:
    if "taxid:9606" in mitab[9]:
        EBV_HUM_prot.add(mitab[0]) #push 1st partner (human)
        EBV_prot.add(mitab[1]) #so second is viral
    else:
        EBV_HUM_prot.add(mitab[1])
        EBV_prot.add(mitab[0])

print(f"Nb EBV prot in EBV - EBV data set is { len(EBV_prot) }")
print(f"Nb EBV prot in EBV - HUM data set is { len(EBV_HUM_prot) }")
```
